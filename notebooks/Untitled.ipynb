```julia
using ControlSystems
```

```julia
theta_1 = 135;
theta_2 = atan2(1,-2)*180/pi
theta_z = atan2(1, -0.5)*180/pi;
theta_p = 180 - theta_1 - theta_2 + theta_z;

p = -1/tan(theta_p*pi/180)

G = tf([1], [1, -1, 0])
F = tf([1, 0.5], [1, -p])
```

```julia
f=rlocus(G*F)
```
