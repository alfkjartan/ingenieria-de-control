# Visualize calculation of the response of a filter from pole and zero locations
A filter $F(s)$ has the zeros $\{q_1, q_2, \ldots, q_m\}$ and poles $\{p_1, p_2, \ldots, p_m\}$
$$ F(s) = k\frac{(s-q_1)(s-q_2)\cdots(s-q_m)}{(s-p_1)(s-p_2)\cdots(s-p_n)}. $$

The gain of the system for a certain point $s$ in s-plane is then given by $k$ times the product of the distances from each zero to the point $s$ divided by the product of the distances from each pole to the point $s$.
$$ |F(s)| = k \frac{|s-q_1||s-q_2|\cdots|s-q_m|}{|s-p_1||s-p_2|\cdots|s-p_n|} $$
