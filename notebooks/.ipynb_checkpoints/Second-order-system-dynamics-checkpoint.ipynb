```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as signal
%matplotlib inline
```

```python
def stepresponse(wn, zeta, T, fig, line):
    # Define the continuous-time linear second order system
    num = [wn**2];
    den = [1, 2*zeta*wn, wn**2]
    G = signal.lti(num, den)

    # Plot a step response
    (t, y) = signal.step(G, T=T)
    
    # Update the plot
    line.set_xdata(t)
    line.set_ydata(y)

    fig.canvas.draw()
```

```python
import time
from IPython import display

N = 400
T = np.linspace(0,10, 400)

fig = plt.figure( figsize=(16,8))
ax = fig.add_subplot(111)
uline, = ax.plot(T, np.ones(N), ':') # Returns a tuple of line objects, thus the comma
line1, = ax.plot(T, np.zeros(N)) # Returns a tuple of line objects, thus the comma
plt.xlim((0, 10))
plt.ylim((0,2))
plt.xlabel("Time [s]")
wn = 4
zetas = [2,1,0.707, 0.5,0.25, 0.125]

for z in zetas:
    stepresponse(wn, z, T, fig, line1)
    display.clear_output(wait=True)
    display.display(plt.gcf())
    time.sleep(1.0)
```
