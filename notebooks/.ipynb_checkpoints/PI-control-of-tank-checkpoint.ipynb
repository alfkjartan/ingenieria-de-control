# Step response of the tank example with PI-control

The closed-loop transfer function is
$$ Y(s) = \frac{4s(s+2)}{s(s+1)(s+2) + 8K_ps + 8K_i}V(s) $$
where $K_p$ is the *proportional gain* and $K_i$ is the *integral gain*. 

```julia
using Interact
using ControlSystems
using Gadfly
```

```julia
N = 600; # Number of samples

t = collect(linspace(0, 10, N)); # 60 values per second 

# The input step
v = ones(N);

@manipulate for Kp in 0:0.1:6, Ki in 0:0.1:12
    S = 4*tf([1.0, 2.0, 0], [1.0, 3.0, 2+8Kp, 8Ki]);
    yr, t_out, x = lsim(S, v, t);
    stp = plot(x=t, y=yr, Geom.line, Guide.xlabel("time [s]"), Guide.ylabel("y"), Theme(line_width=2pt))
end
```
