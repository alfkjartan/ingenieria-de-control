```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as signal
%matplotlib inline
```

```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
```

```python
def stepresponse(wn, zeta, fig, line):
    # Define the continuous-time linear second order system
    num = [wn**2];
    den = [1, 2*zeta*wn, wn**2]
    G = signal.lti(num, den)

    # Plot a step response
    (t, y) = signal.step(G)
    
    # Update the plot
    line.set_xdata(t)
    line.set_ydata(y)

    fig.canvas.draw()
```

```python
fig = plt.figure( figsize=(16,8))
ax = fig.add_subplot(111)
uline, = ax.plot(np.linspace(0,10,300), np.ones(300), ':') # Returns a tuple of line objects, thus the comma
line1, = ax.plot(np.linspace(0,10,300), np.zeros(300)) # Returns a tuple of line objects, thus the comma
plt.xlim((0, 10))
plt.ylim((0,2))
plt.xlabel("Time [s]")
wn = 1
zeta = 0.5

interact(stepresponse, wn=2, zeta=0.4, fig=fixed(fig), line=fixed(line1))
```
