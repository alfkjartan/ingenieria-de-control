# Effect of zero on the step-response of a second-order system

```python
import numpy as np
import control.matlab as cm
import re
import matplotlib.pyplot as plt
from scipy import signal
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
def second_order_step(z, line, zline, fig, zeta, wn):
    if (z != 0):
        G = cm.tf([-1.0/z, 1]*wn*wn, [1, 2*zeta*wn, wn*wn])
        zline.set_xdata([z])
        y, t = cm.step(G)
        line.set_xdata(t)
        line.set_ydata(y)
        fig.canvas.draw()
        fig.canvas.flush_events()
  
```

```python
%matplotlib notebook
fig = plt.figure(figsize=(8,5))
ax1 = fig.add_subplot(121)
ax1.plot([-0.8, -0.8], np.sqrt(1-0.8**2)*np.array([1, -1]), 'x', markersize=12)
zline, = ax1.plot([-2], [0], 'o', markersize=12)
ax1.plot([-3,1], [0, 0], 'k')
ax1.plot([0,0], [-2, 2], 'k')
plt.ylim((-2,2))
plt.xlim((-3,1))
ax1.grid(False)

ax2 = fig.add_subplot(122)
line, = ax2.plot([0, 9], [0, 0], linewidth=2)
plt.ylim((-1,2))
plt.xlim((0,9))
plt.show()
```

```python
interact(second_order_step, z=(-4.0, 2.0), 
                               line=fixed(line), zline=fixed(zline),fig=fixed(fig), 
                               zeta=fixed(0.8), wn=fixed(1))
```
